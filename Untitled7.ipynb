1. Implement a basic Statistical Machine Translation (SMT) model that uses word-by-word translation with a dictionary lookup approach.



```python
# Simple Dictionary-based SMT model using a word-by-word translation approach
# Note: This is a very basic and naive implementation for educational purposes.

# Sample English to French translation dictionary
translation_dict = {
    'hello': 'bonjour',
    'world': 'monde',
    'good': 'bon',
    'morning': 'matin'
}

def translate(sentence, dictionary):
    words = sentence.split()
    translated_words = [dictionary.get(word, word) for word in words]  # Default to original word if not found
    return ' '.join(translated_words)

# Example Usage
english_sentence = "hello world"
french_sentence = translate(english_sentence, translation_dict)
print(f"Translated sentence: {french_sentence}")
```

 Implement an Attention mechanism in a Neural Machine Translation (NMT) model using PyTorch.

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Simple Attention Mechanism in NMT using PyTorch
class Attention(nn.Module):
    def __init__(self, hidden_size):
        super(Attention, self).__init__()
        self.attn = nn.Linear(hidden_size, hidden_size)
        self.v = nn.Parameter(torch.rand(hidden_size))

    def forward(self, encoder_outputs, decoder_hidden):
        attn_weights = torch.matmul(encoder_outputs, decoder_hidden.unsqueeze(2))  # Compute attention weights
        attn_weights = torch.nn.functional.softmax(attn_weights.squeeze(2), dim=1)  # Apply softmax
        context = torch.matmul(attn_weights.unsqueeze(1), encoder_outputs).squeeze(1)  # Context vector
        return context, attn_weights

# Example Encoder-Decoder setup with attention mechanism
class NMTModel(nn.Module):
    def __init__(self, vocab_size, hidden_size):
        super(NMTModel, self).__init__()
        self.encoder = nn.LSTM(vocab_size, hidden_size, batch_first=True)
        self.attention = Attention(hidden_size)
        self.decoder = nn.LSTMCell(hidden_size, hidden_size)
        self.output_layer = nn.Linear(hidden_size, vocab_size)

    def forward(self, input_seq):
        encoder_outputs, (h_n, c_n) = self.encoder(input_seq)
        context, _ = self.attention(encoder_outputs, h_n)
        decoder_output, (h, c) = self.decoder(context)
        output = self.output_layer(decoder_output)
        return output

# Example usage: dummy input
nmt_model = NMTModel(vocab_size=10, hidden_size=16)
input_seq = torch.rand(1, 5, 10)  # Dummy input: batch size 1, sequence length 5, vocab size 10
output = nmt_model(input_seq)
print(output)
```

Use a pre-trained GPT model to perform machine translation from English to French.


```python
from transformers import MarianMTModel, MarianTokenizer

# Load a pre-trained MarianMT model for English-to-French translation
model_name = 'Helsinki-NLP/opus-mt-en-fr'
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)

def translate_to_french(text):
    inputs = tokenizer.encode(text, return_tensors="pt")
    translated = model.generate(inputs, max_length=512)
    translation = tokenizer.decode(translated[0], skip_special_tokens=True)
    return translation

# Example usage
english_text = "Hello, how are you?"
french_translation = translate_to_french(english_text)
print(f"French translation: {french_translation}")
```

Generate a short poem using GPT-2 for a specific theme (e.g., "Nature").

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load a pre-trained GPT-2 model
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

def generate_poem(theme):
    prompt = f"A short poem about {theme}:"
    inputs = tokenizer.encode(prompt, return_tensors="pt")
    outputs = model.generate(inputs, max_length=100, num_return_sequences=1, no_repeat_ngram_size=2)
    poem = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return poem

# Example usage
theme = "Nature"
poem = generate_poem(theme)
print(poem)
```

 Implement a basic reinforcement learning setup for text generation using PyTorch's reward function.

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define a simple reward-based text generation model
class TextGenRL(nn.Module):
    def __init__(self, vocab_size, hidden_size):
        super(TextGenRL, self).__init__()
        self.lstm = nn.LSTM(vocab_size, hidden_size)
        self.fc = nn.Linear(hidden_size, vocab_size)

    def forward(self, input_seq):
        lstm_out, _ = self.lstm(input_seq)
        logits = self.fc(lstm_out)
        return logits

# Reward function (simple)
def reward_function(generated_text, target_text):
    # Reward based on similarity (e.g., matching words)
    return torch.sum(generated_text == target_text).item()

# Example usage
model = TextGenRL(vocab_size=50, hidden_size=128)
optimizer = optim.Adam(model.parameters(), lr=0.001)
input_seq = torch.randn(1, 5, 50)  # Dummy input
target_text = torch.randint(0, 50, (1, 5))  # Dummy target

# Forward pass
logits = model(input_seq)

# Compute reward (just a simple comparison here)
reward = reward_function(logits, target_text)
print(f"Reward: {reward}")
```

 Create a simple multimodal generative model that generates an image caption given an image.


```python
from transformers import VisionEncoderDecoderModel, ViTImageProcessor, AutoTokenizer

# Load pre-trained multimodal model for image captioning
model = VisionEncoderDecoderModel.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
processor = ViTImageProcessor.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
tokenizer = AutoTokenizer.from_pretrained("gpt2")

def generate_caption(image_path):
    # Load and process image
    image = processor(images=image_path, return_tensors="pt").pixel_values
    # Generate caption
    output_ids = model.generate(image)
    caption = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return caption

# Example usage: provide an image file path
image_path = "path_to_image.jpg"
caption = generate_caption(image_path)
print(f"Generated Caption: {caption}")
```

 Demonstrate how to evaluate bias in generated content by analyzing GPT responses to prompts with potentially sensitive terms.


```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load GPT-2 model
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Analyze GPT-2 responses to sensitive prompts
def analyze_bias(prompt):
    inputs = tokenizer.encode(prompt, return_tensors="pt")
    outputs = model.generate(inputs, max_length=50, num_return_sequences=1)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text

# Example usage: potentially sensitive prompt
prompt = "What do you think about gender roles?"
generated_content = analyze_bias(prompt)
print(f"Generated content: {generated_content}")
```

 Create a simple Neural Machine Translation model with PyTorch for translating English phrases to German.


```python
import torch
import torch.nn as nn
import torch.optim as optim

class NMTModel(nn.Module):
    def __init__(self, input_vocab_size, output_vocab_size, embedding_dim, hidden_dim):
        super(NMTModel, self).__init__()
        self.embedding = nn.Embedding(input_vocab_size, embedding_dim)
        self.rnn = nn.LSTM(embedding_dim, hidden_dim)
        self.fc = nn.Linear(hidden_dim, output_vocab_size)

    def forward(self, input_seq):
        embedded = self.embedding(input_seq)
        output, (h_n, c_n) = self.rnn(embedded)
        output = self.fc(output)
        return output

# Example: Simple English-to-German translation model
input_vocab_size = 100  # Dummy vocab size
output_vocab_size = 100
embedding_dim = 64
hidden_dim = 128

nmt_model = NMTModel(input_vocab_size, output_vocab_size, embedding_dim, hidden_dim)

# Dummy input (English) and target (German)
input_seq = torch.randint(0, input_vocab_size, (1, 5))  # Batch size 1, seq length 5
output_seq = nmt_model(input_seq)
print(output_seq)
```
